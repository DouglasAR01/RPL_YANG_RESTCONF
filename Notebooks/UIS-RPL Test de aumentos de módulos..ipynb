```python
router.routing.router_id = '1.1.1.1'
router.routing.interfaces.interface.append('wireless1')
```

```python
rpl = router.routing.control_plane_protocols.control_plane_protocol.add(
    type = 'rpl',
    name = 'RPL'
)
```

```python
rpl.rpl.trickle.i_min = 60*1000 # 1 minute to milliseconds
rpl.rpl.trickle.i_max = 4
rpl.rpl.trickle.k = 2
rpl.rpl.interfaces.interface.add('wireless1')

# OBJECTIVE FUNCTION
of = rpl.rpl.objective_functions.objective_function.add(
    type = 'of-of0',
    name = 'of0'
)

# NEIGHBOR 1
# Cost is ETX
n1 = rpl.rpl.routing_addresing.upward_routing.neighbors.neighbor.add(
    ipv6_address = '2002:0DB8:AC10:FE01::'
)
n1._set_last_update('2020-03-04T14:20:50.52Z')
n1._set_rank(1)
n1._set_cost(1)

#NODE 1
npp1 = rpl.rpl.routing_addresing.upward_routing.node_preferred_parents.node_preferred_parent.add(
    ipv6_address = n1.ipv6_address
)
npp1._set_parent_ipv6_address('2001:0DB8:AC10:FE01::')
npp1._set_parent_rank(0)
npp1._set_parent_cost(1)
npp1._set_rank(n1.rank)
npp1._set_cost(n1.cost)

#NODE 2
npp2 = rpl.rpl.routing_addresing.upward_routing.node_preferred_parents.node_preferred_parent.add(
    ipv6_address = '2003:0DB8:AC10:FE01::'
)
npp2._set_parent_ipv6_address(npp1.ipv6_address)
npp2._set_parent_rank(1)
npp2._set_parent_cost(1.3)
npp2._set_rank(2)
npp2._set_cost(2.3)

#DODAG TOPOLOGY OPERATIONAL STATE DATA POPULATION
dt = rpl.rpl.routing_addresing.upward_routing.dodag_topology
dt._set_objective_function(of.name)
dt._set_num_of_nodes(3)
dt.dodag_version._set_rpl_instance_id(1)
dt.dodag_version._set_dodag_id('fd12:3456:789a:1::1')
dt.dodag_version._set_dodag_version_number(2)
```
